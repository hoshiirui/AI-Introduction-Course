```python
import networkx as nx
import matplotlib.pyplot as plt

edge = []
jmledge = input("Masukkan jumlah edge: ")
for _ in range(int(jmledge)):
  x = [int(x) for x in input("Masukkan angka (node1 node2 bobot): ").split()]
  edge.append(x)

print(edge)

# -------------- ADJACENCY MATRIX ----------------
n = max(max(x[:2]) for x in edge)
adj_matrix = [[0 for i in range(int(n))] for j in range(int(n))]

for i in edge:
    node1, node2, weight = i
    adj_matrix[node1-1][node2-1] = weight
    adj_matrix[node2-1][node1-1] = weight

# Print the adjacency matrix
print("Adjacency Matrix:")
for row in adj_matrix:
    print(row)

# ------------ ADJACENCY LIST ----------------

adj_list = {}

# masukin
for i in edge:
    node1, node2, bobot = i
    if node1 not in adj_list:
        adj_list[node1] = [(node2, weight)]
    else:
        adj_list[node1].append((node2, weight))

    if node2 not in adj_list:
        adj_list[node2] = [(node1, weight)]
    else:
        adj_list[node2].append((node1, weight))

# Print the adjacency list
print("Adjacency List:")
print(adj_list)

# -------------- Represent Using Adjacency Matrix ---------------
print("Graph Representation:")
G = nx.Graph()
G.add_nodes_from(range(1, int(n)))
for i in edge:
    G.add_edge(i[0], i[1], weight=i[2])

pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True)
edge_labels = nx.get_edge_attributes(G, 'weight')
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels)
plt.show()
```

```python
#bfs
def bfs(adj_matrix, start_node):
    
    visited = [False for i in range(n)]
    queue = [start_node]
    visited[start_node] = True

    while queue:
        node = queue.pop(0)
        print(node + 1, end=' ')

        for neighbor, weight in enumerate(adj_matrix[node]):
            if weight and not visited[neighbor]:
                visited[neighbor] = True
                queue.append(neighbor)

#dfs
def dfs(graph, start_node):
    visited = set()
    stack = [start_node]
    while stack:
        node = stack.pop()
        if node not in visited:
            visited.add(node)
            print(node + 1, end=' ')
            for neighbor, weight in enumerate(graph[node]):
                if weight != 0 and neighbor not in visited:
                    stack.append(neighbor)


print("Visit order (bfs): ")
bfs(adj_matrix, 0)

print("\nVisit order (bfs): ")
dfs(adj_matrix, 0)
```
